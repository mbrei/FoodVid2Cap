# Generate Plots of training loss and training and validations scores during training
During training the loss and several evaluation metrics that measure the model's captioning capabilities (including the BLEU (1-4) metricand the METEOR score) are calculated for the training and validation set. The developement of these scores over the epchos are plotted in the following

```python
from ast import literal_eval 
import matplotlib.pyplot as plt
```

```python
def plot_curves(train,val,title, save_path): 
    
    plt.subplot(2, 1, 1)
    plt.title(title)
    plt.plot(train, '-',label='')
    plt.plot(val, '-',label='validation')
    plt.xlabel('Epoch')
    plt.legend(loc='lower right')
    plt.gcf().set_size_inches(15, 12)
    plt.savefig(save_path + title +'.png',transparent=True)
    plt.show()
```

#### Specify the model path and the path where the plots should be saved to 

```python
model_result_path = "../models/2019-08-02_17-12-01/"
save_path = "../plots/"
```

#### Training Loss Curve

```python
with open(model_result_path + "loss_train_list.json", 'r') as f:
    losses_train = literal_eval(f.read())

plt.subplot(2, 1, 1)
plt.title("Training Loss Curve",fontsize = 18)
plt.plot(losses_train, '-',label='Zero-Shot')
plt.xlabel('Epoch', fontsize = 14)
plt.legend(loc='upper right', fontsize = 17)
plt.gcf().set_size_inches(15, 12)
plt.savefig(save_path + 'loss_zeroShot_baseline.png',transparent=True)
plt.show()
```

#### Developement of the METEOR score over the epochs for the training and validation set

```python
with open(model_result_path + "meteor_train_list.json", 'r') as f:
    meteor_train = literal_eval(f.read())
with open(model_result_path + "meteor_val_list.json", 'r') as f:
    meteor_val = literal_eval(f.read())

plot_curves(meteor_train,meteor_val,"METEOR scores", save_path)
```

#### Developement of the BLEU1 score over the epochs for the training and validation set

```python
with open(model_result_path + "bleus1_train_list.json", 'r') as f:
    bleu1_train = literal_eval(f.read())
with open(model_result_path + "bleus1_val_list.json", 'r') as f:
    bleu1_val = literal_eval(f.read())
    
plot_curves(bleu1_train,bleu1_val,"BLEU1 scores", save_path)
```
